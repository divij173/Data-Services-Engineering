```python
import pandas as pd

#question 1
colnames=['_id', 'Month', 'AustralianPort', 'ForeignPort', 'Country', 'Passengers_In', 'Freight_In_(tonnes)', 'Mail_In_(tonnes)', 'Passengers_Out', 'Freight_Out_(tonnes)', 'Mail_Out_(tonnes)', 'Passengers_Total', 'Freight_Total_(tonnes)', 'Mail_Total_(tonnes)', 'Year', 'Month_num'] 
df = pd.read_csv('city_pairs.csv', skiprows = 1, names=colnames, header=None)


def passengers_val(df):
    if (df['Passengers_In'] < df['Passengers_Out']):
        return 'OUT'
    elif (df['Passengers_In'] > df['Passengers_Out']):
        return 'IN'
    elif (df['Passengers_In'] == df['Passengers_Out']):
        return 'SAME'
    
def freight_val(df):
    if (df['Freight_In_(tonnes)'] < df['Freight_Out_(tonnes)']):
        return 'OUT'
    elif (df['Freight_In_(tonnes)'] > df['Freight_Out_(tonnes)']):
        return 'IN'
    elif (df['Freight_In_(tonnes)'] == df['Freight_Out_(tonnes)']):
        return 'SAME'
    
def mail_val(df):
    if (df['Mail_In_(tonnes)'] < df['Mail_Out_(tonnes)']):
        return 'OUT'
    elif (df['Mail_In_(tonnes)'] > df['Mail_Out_(tonnes)']):
        return 'IN'
    elif (df['Mail_In_(tonnes)'] == df['Mail_Out_(tonnes)']):
        return 'SAME'
    
df['passenger_in_out'] = df.apply(passengers_val, axis = 1)
df['freight_in_out'] = df.apply(freight_val, axis = 1)
df['mail_in_out'] = df.apply(mail_val, axis = 1)

df.head()
```

```python
#question 2

df1 = df[['AustralianPort', 'Month', 'passenger_in_out', 'freight_in_out', 'mail_in_out']]
df1 = pd.get_dummies(df1, columns=['passenger_in_out', 'freight_in_out', 'mail_in_out'])
df1 = df1[df1['Month'] >= 'Jan-85']
df1 = df1[['AustralianPort', 'Month', 'passenger_in_out_IN', 'passenger_in_out_OUT', 'freight_in_out_IN', 'freight_in_out_OUT', 'mail_in_out_IN', 'mail_in_out_OUT']]

df1 = df1.groupby('AustralianPort').sum()
df1 = df1.sort_values('passenger_in_out_IN', ascending=False)

df1.columns = ['PassengerInCount', 'PassengerOutCount', 'FreightInCount', 'FreightOutCount', 'MailInCount', 'MailOutCount']
df1 = df1.reset_index(drop=True)
#print(df1)
df1.head()
```

```python
#question 3

df2 = df[['AustralianPort', 'Month', 'Country', 'Passengers_In', 'Passengers_Out', 'Freight_In_(tonnes)', 'Freight_Out_(tonnes)', 'Mail_In_(tonnes)', 'Mail_Out_(tonnes)']] 
df2 = df2.groupby(['Country', 'Month']).sum()
#df2 = df2.reset_index()
df2 = df2.groupby('Country').mean()
df2.columns = ['Passengers_in_average', 'Passengers_out_average', 'Freight_in_average', 'Freight_out_average', 'Mail_in_average', 'Mail_out_average']
df2.head()
```

```python
#question 4
df3 = df[['AustralianPort', 'Month', 'Country', 'Passengers_Out']]
df3 = df3.groupby(['Country', 'Month', 'AustralianPort']).count().reset_index()

df3 = df3[df3['Passengers_Out'] != 1]
df3['Passengers_Out'] = 1
df3 = df3.groupby(['Country']).sum()
df3 = df3.sort_values('Passengers_Out', ascending = False)
df3.columns = ['Unique_ForeignPort_Count']

df3.head()
```

```python
#question 5

df5 = pd.read_csv('seats.csv')

def passengers_source(df5):
    if (df5['In_Out'] == 'I'):
        return df5['International_City']
    elif (df5['In_Out'] == 'O'):
        return df5['Australian_City']
    
def passengers_destination(df5):
    if (df5['In_Out'] == 'I'):
        return df5['Australian_City']
    elif (df5['In_Out'] == 'O'):
        return df5['International_City']
    
df5['Source_City'] = df5.apply(passengers_source, axis=1)
df5['Destination_City'] = df5.apply(passengers_destination, axis=1)

df5.head()
```
